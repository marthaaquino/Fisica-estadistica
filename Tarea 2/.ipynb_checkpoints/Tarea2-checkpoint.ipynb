Creando una funciÃ³n que obtenga los primeros 4 momentos de un conjunto de datos.

```julia
function momentos(datos)
    #media de los datos
    Î¼=mean(datos)
    #Areglos donde se indexarÃ¡n (x_i-<x>)^n
    variables_2=[]
    variables_3=[]
    variables_4=[]
    for i in 1:length(datos)
        push!(variables_4,(datos[i]-Î¼)^4)
        push!(variables_3,(datos[i]-Î¼)^3)
        push!(variables_2,(datos[i]-Î¼)^2)
    end
    return[0,mean(variables_2),mean(variables_3),mean(variables_4)]
end
```

Calculando los momentos de los valores.

```julia
tabla=DataFrame(Momento=["Momento 1","Momento 2","Momento 3","Momento 4"],
Distribucion1=momentos(Distribucion1),
Distribucion2=momentos(Distribucion2),
Distribucion3=momentos(Distribucion3)
)
```
