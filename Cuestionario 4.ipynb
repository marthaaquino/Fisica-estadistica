##### Ejercicio 1:
Supón que tienes un fluido con vicosidad $\gamma$ y coeficiente de fluctuación $\beta$.
a) Esquemáticamente, muestra cómo cambia la probabilidad de que una partícula en dicho sistema tenga velocidad $v$ como función del tiempo.

De clase sabemos que la distribución de probabilidad es 
$$p(v,t)=\exp\left(-\frac{(v-v_0e^{-\gamma t})^2}{(2\beta^2/2\gamma(1-e^{-2\gamma t})}\right){\frac{1}{\sqrt{2\pi\left({\frac{\beta^2}{2\gamma}} \right)(1-\exp(-2\gamma t))}}} $$

Creamos una función que tome a β y γ como parametros

```julia
function p(γ,β,v,vo,tiempo)
    pv=[]
   for t in tiempo
        a=exp(-2*γ*t)
        b=exp(-(v-vo*a)^2/(β^2/γ*(1-a)))
        c=2*3.1416*((β^2)/(2*γ))*(1-a)
        d=b*(c)^(-1/2)
        push!(pv,d)
    end
    return pv
end
```

Para darnos una idea consideremos β=1, γ=1, $v_0=0$ y $t_0=0$. Graficando la evolución temporal de la probabilidad para diferentes velocidades iniciales.


```julia
#Definimos el rango de tiempo
β=1
γ=1
v0=0
t=collect(0:0.01:5)
#El conjunto de velocidades iniciales.
v=[0,0.1,0.5,1,5,10,-0,-0.1,-0.5,-1,5,-10]
for i in v
    plot!(t,p(γ,β,i,v0,t),label="Velocidad = $(i)")
end
plot!(title="p(v,t)",x_label="t",y_label="p(t)")
```
