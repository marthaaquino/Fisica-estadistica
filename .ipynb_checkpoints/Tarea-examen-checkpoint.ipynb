```julia
function FU(x,kbT)
    #s=((1-x)/2)*log((1-x)/(2))+((1+x)/2)*log((1+x)/2)
    resultado=zeros(length(x))
    for i in 1:length(x)
        resultado[i]=-x[i]*x[i]+kbT*( ((1-x[i])/2)*log((1-x[i])/(2))+((1+x[i])/2)*log((1+x[i])/2))
    end
    return resultado
end
```

```julia
kbT_coleccion=[0.1,1,1.3,2,3]
x_coleccion=collect(-1:0.01:1);
```

```julia
for i in kbT_coleccion
    plot!(x_coleccion,FU(x_coleccion,i),label="kbT=$(i)",title="F/U_0",xlabel="x",ylabel="F/U0")
end
plot!(title="F/U_0",xlabel="x",ylabel="F/U0")
```
