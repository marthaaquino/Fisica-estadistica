Lo que hiciste en el punto 3 fue obtener 10 realizaciones de la variable aleatoria 
$\overline{X}=(X1+X2+...+Xn)/n$, donde cada una de las variables $X_i$ que la constituyen es idÃ©ntica a aquella que construiste en el inciso 1. Nota que como discutimos en clase,  $\overline{X}$ no es el promedio de una de las $X_i$'s, sino una variable aleatoria cuyo promedio es igual al promedio de $<Xi>$.
```julia
#Creamos una listas que tendrÃ¡ el promedio de cada una de las 10 realizaciones anteriores
promedios=[]
#Creamos un ciclo para obtener el promedio de cada realizaciÃ³n
for i in 1:10
    #Definimos una suma donde iremos sumando :p los elementos de cada realizaciÃ³n
    suma=0
    for j in 1:10
        suma+=listas[i][j]
    end
    #Obtenemos el promedio de la i-Ã©sima realizaciÃ³n y lo anexamos a la lista de promedios
    promedio=suma/10
    push!(promedios,promedio)
end
```

#### Ejercicio 4. 
(Nota: El cÃ³digo debe tener como variables a n y L). Construye ahora L = 10, 000
listas distintas para cada uno de los siguientes valores de n: n = {10, 100, 1000, 10000}. (Â¡No las reportes!; n son el nÃºmero de realizaciones o pasos). GrÃ¡fica el promedio de cada una de las listas como funciÃ³n del nÃºmero de lista, fijando la escala de los ejes en [-1,1] en todas las grÃ¡ficas.

```julia
#Creamos una funciÃ³n que genere una caminata de n elementos 
function realizacion(n)
    #La lista donde se anexarÃ³n los pasos 
   lista=[]
    #Un ciclo que cree n pasos de la realizaciÃ³n
    for i in 1:n
        push!(lista,rand((-1,1)))
    end
    #Se regresa la lista con los pasos
    return lista
end
```

Construyendo las L listas para n=10,100,1000,10000

```julia
#Las variables: los elementos que puede tomar n
n=[10,100,1000,10000]
#El nÃºmero de listas para cada n
L=10000
#Creamos una lista donde anexaremos las realizaciones
Listas_4=[]
#Para saber a que elemento de n corresponde la lista_4-i creamos un arreglo de ns
ns=[]
#Para cada n dada creamos 10000 listas:
for i in n
    for j in 1:L
        #anexamos esas listas a Listas_4 y anexamos tambiÃ©n la n correspondiente a cada n 
        push!(Listas_4,realizacion(i))
        push!(ns,i)
    end
end
```

Creamos una funciÃ³n que permita obtener el promedio de una lista:

```julia
function promedio(lista)
    #variable de suma
    suma=0
    #Ciclo que recorra todos los elementos de la lista
    for i in 1:length(lista)
        #Suma el i-esimo elemento de la lista a la variable suma
        suma+=lista[i]
    end
    #Dividimos la suma entre el nÃºmero de elementos para obtener el promedio y regresa este valor
    return suma/length(lista)
end
```

```julia
#Obteniendo el promedio de cada lista
promedios_4=[]
for i in Listas_4
    push!(promedios_4,promedio(i))
end
```

```julia
#Creamos listas que agrupen los promedios considerando el nÃºmero de pasos correspondiente a ese promedio
promedio_10=[]
promedio_100=[]
promedio_1000=[]
promedio_10000=[]
#Creamos un ciclo que pase por todos los elementos de los promedios que es la misma longitud que la Lista_4
for i in 1:length(Listas_4)
    #Creamos condicionales para agrupar las listas dependiendo del nÃºmero de n que tienen.
    if ns[i]==10
        push!(promedio_10,promedios_4[i])
    end
    if ns[i]==100
        push!(promedio_100,promedios_4[i])
    end
    if ns[i]==1000
        push!(promedio_1000,promedios_4[i])
    end
    if ns[i]==10000
        push!(promedio_10000,promedios_4[i])
    end
end
```

Graficando para cada n y nombrando X_n a las listas con n pasos.

```julia
#La grÃ¡fica para n=10
plot(promedio_10,ylabel="Promedio de X", xlabel="NÃºmero de lista", 
                title="n=10",label= false, color="#65ADC6")
```

a. Obteniendo el promedio de los promedios, para ello creamos una funciÃ³n que obtenga el promedio de los promedios:

```julia
function promedio_promedios(lista)
    #Creamos una lista donde almacenaremos los promedios del promedio
    promedios=[]
    for i in 1:length(lista)
        #obtenemos el promedio de cada lista y lo almacenamos en la lista promedios
        push!(promedios,mean(lista[i]))
    end
    #Obtenemos el promedio de los promedios y lo regresamos con la funciÃ³n
return mean(promedios)
end
```

Presentamos los promedios en una tabla para mejor visualizaciÃ³n

b. Calculando las varianzas de los promedios usando el mismo procedimiento que para los promedios de los promedios

```julia
#Creamos la funciÃ³n que nos cree la varianza de la varianza de una lista
function varianza_promedio(lista)
    #Creamos una variable que almacenarÃ¡ la vrianza de la iesima realizaciÃ³n de la lista
    promedio=[]
    for i in 1:length(lista)
        push!(promedio,mean(lista[i]))
    end
    #Regresamos la varianza de las varianza
return var(promedio,corrected=false)
end
```

Calculando las varianza de los promedios

```julia
#Obteniendo la varianza de las varianzas
varianza_promedio_10=varianza_promedio(listas_10)
varianza_promedio_100=varianza_promedio(listas_100)
varianza_promedio_1000=varianza_promedio(listas_1000)
varianza_promedio_10000=varianza_promedio(listas_10000)
DataFrame(n10=varianza_promedio_10,n100=varianza_promedio_100,
                m1000=varianza_promedio_1000,n10000=varianza_promedio_10000)
```

Graficando la varianza de los promedios en escala logaritmica en ambos ejes.

```julia
varianza_promedio_n=[varianza_promedio_10,varianza_promedio_100,varianza_promedio_1000,varianza_promedio_10000]
plot(n,varianza_promedio_n,xscale=:log10,yscale=:log10,ylabel="log (VAR{<X>})", 
    xlabel="log(n)",label=false,color="#13165B", 
        marker=:star,title="log(VAR{<X>}) en funciÃ³n de n")
```
