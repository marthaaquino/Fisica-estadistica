#### Ejercicio 4. 
(Nota: El código debe tener como variables a n y L). Construye ahora L = 10, 000
listas distintas para cada uno de los siguientes valores de n: n = {10, 100, 1000, 10000}. (¡No las reportes!; n son el número de realizaciones o pasos). Gráfica el promedio de cada una de las listas como función del número de lista, fijando la escala de los ejes en [-1,1] en todas las gráficas.

```julia
#Creamos una función que genere una lista de n elementos 
function realizacion(n)
   lista=[]
    for i in 1:n
        push!(lista,rand((-1,1)))
    end
    return lista
end
```

Construyendo las L listas para n=10,100,1000,10000

```julia
n=[10,100,1000,10000]
L=10000

Listas_4=[]
ns=[]

for i in n
    for j in 1:L
        push!(Listas_4,realizacion(i))
        push!(ns,i)
    end
end
```

Creando una función que obtenga el promedio de una lista 

```julia
function promedio(lista)
    suma=0
    for i in 1:length(lista)
        suma+=lista[i]
    end
    return suma/length(lista)
end
```

```julia
#Obteniendo el promedio de cada lista
promedios_4=[]
for i in Listas_4
    push!(promedios_4,promedio(i))
end
```

Graficando

```julia
promedio_10=[]
promedio_100=[]
promedio_1000=[]
promedio_10000=[]
for i in 1:length(Listas_4)
    if ns[i]==10
        push!(promedio_10,promedios_4[i])
    end
    if ns[i]==100
        push!(promedio_100,promedios_4[i])
    end
    if ns[i]==1000
        push!(promedio_1000,promedios_4[i])
    end
    if ns[i]==10000
        push!(promedio_10000,promedios_4[i])
    end
end
```

```julia
#La gráfica para n=10
plot(promedio_10,ylabel="Promedio de la variable estocastica", xlabel="Realización (o número de lista)", title="n=10",label= false)
```

```julia
#La gráfica para n=10
plot(promedio_100,ylabel="Promedio de la variable estocastica", xlabel="Realización (o número de lista)", title="n=100",label=false)
```

a. Obteniendo el promedio de los promedios, para ello creamos una función que obtenga el promedio de los promedios:

```julia
function promedio_promedios(lista)
    promedios=[]
    for i in 1:length(lista)
        push!(promedios,mean(lista[i]))
    end
return mean(promedios)
end
```

```julia
#Obteniendo los promedios de los promedios
promedio_promedio_10=promedio_promedios(listas_10)
promedio_promedio_100=promedio_promedios(listas_100)
promedio_promedio_1000=promedio_promedios(listas_1000)
promedio_promedio_10000=promedio_promedios(listas_10000)
promedio_promedio=[promedio_promedio_10,promedio_promedio_100,promedio_promedio_1000,promedio_promedio_10000]
```

b. Calculando las varianza usando el mismo procedimiento

```julia
function varianza_varianzas(lista)
    varianza=[]
    for i in 1:length(lista)
        push!(varianza,var(lista[i]))
    end
return mean(varianza)
end
```

```julia
#Obteniendo la varianza de las varianzas
varianza_varianza_10=varianza_varianzas(listas_10)
varianza_varianza_100=varianza_varianzas(listas_100)
varianza_varianza_1000=varianza_varianzas(listas_1000)
varianza_varianza_10000=varianza_varianzas(listas_10000)
varianza_varianza_n=[varianza_varianza_10,varianza_varianza_100,varianza_varianza_1000,varianza_varianza_10000]
```

```julia
plot(n,varianza_varianza_n,xscale=:log10,yscale=:log10,ylabel="Varianza de la varianza (escala log)", xlabel="log(n)",label=false)
```
