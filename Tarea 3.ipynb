GrÃ¡fica el promedio de cada una de las listas como funciÃ³n del nÃºmero de lista, fijando la escala de los ejes en [-1,1] en todas las grÃ¡ficas.

```julia
#Creamos una funciÃ³n que genere una realizaciÃ³n de caminante de n elementos 
function realizacion(n)
    #La lista donde se anexarÃ³n los pasos 
   lista=[]
    #Un ciclo que cree n pasos de la realizaciÃ³n
    for i in 1:n
        push!(lista,rand((-1,1)))
    end
    #Se regresa la lista con los pasos
    return lista
end
```

Construyendo las L listas para n=10,100,1000,10000

```julia
#Las variables: los elementos que puede tomar n
n=[10,100,1000,10000]
#El nÃºmero de listas para cada n
L=10000
#Creamos una lista donde anexaremos las realizaciones
Listas_4=[]
#Para saber a que elemento de n corresponde la lista_4-i creamos un arreglo de ns
ns=[]
#Para cada n dada creamos 10000 listas:
for i in n
    for j in 1:L
        #anexamos esas listas a Listas_4 y anexamos tambiÃ©n la n correspondiente a cada n 
        push!(Listas_4,realizacion(i))
        push!(ns,i)
    end
end
```

Creamos una funciÃ³n que permita obtener el promedio de una lista:

```julia
function promedio(lista)
    #variable de suma
    suma=0
    #Ciclo que recorra todos los elementos de la lista
    for i in 1:length(lista)
        #Suma el i-esimo elemento de la lista a la variable suma
        suma+=lista[i]
    end
    #Dividimos la suma entre el nÃºmero de elementos para obtener el promedio y regresa este valor
    return suma/length(lista)
end
```

```julia
#Obteniendo el promedio de cada lista
promedios_4=[]
for i in Listas_4
    push!(promedios_4,promedio(i))
end
```

```julia
#Creamos listas que anexe el promedio considerando el nÃºmero de n que se tiene
promedio_10=[]
promedio_100=[]
promedio_1000=[]
promedio_10000=[]
#Creamos un ciclo que pase por todos los elementos de los promedios que es la misma longitud que la Lista_4
for i in 1:length(Listas_4)
    #Creamos condicionales para agrupar las listas dependiendo del nÃºmero de n que tienen.
    if ns[i]==10
        push!(promedio_10,promedios_4[i])
    end
    if ns[i]==100
        push!(promedio_100,promedios_4[i])
    end
    if ns[i]==1000
        push!(promedio_1000,promedios_4[i])
    end
    if ns[i]==10000
        push!(promedio_10000,promedios_4[i])
    end
end
```

a. Obteniendo el promedio de los promedios, para ello creamos una funciÃ³n que obtenga el promedio de los promedios:

```julia
function promedio_promedios(lista)
    #Creamos una lista donde almacenaremos los promedios del promedio
    promedios=[]
    for i in 1:length(lista)
        #obtenemos el promedio de cada lista y lo almacenamos en la lista promedios
        push!(promedios,mean(lista[i]))
    end
    #Obtenemos el promedio de los promedios y lo regresamos con la funciÃ³n
return mean(promedios)
end
```

b. Calculando las varianzas de la varianza usando el mismo procedimiento que para los promedios de los promedios

```julia
#Creamos la funciÃ³n que nos cree la varianza de la varianza de una lista
function varianza_varianzas(lista)
    #Creamos una variable que almacenarÃ¡ la vrianza de la iesima realizaciÃ³n de la lista
    varianza=[]
    for i in 1:length(lista)
        push!(varianza,var(lista[i]))
    end
    #Regresamos la varianza de las varianza
return var(varianza)
end
```

Calculando la varianza de las varianzas

```julia
#Obteniendo la varianza de las varianzas
varianza_varianza_10=varianza_varianzas(listas_10)
varianza_varianza_100=varianza_varianzas(listas_100)
varianza_varianza_1000=varianza_varianzas(listas_1000)
varianza_varianza_10000=varianza_varianzas(listas_10000)
DataFrame(n10=varianza_varianza_10,n100=varianza_varianza_100,
                m1000=varianza_varianza_1000,n10000=varianza_varianza_10000)
```

```julia
varianza_varianza_n=[varianza_varianza_10,varianza_varianza_100,varianza_varianza_1000,varianza_varianza_10000]
plot(n,varianza_varianza_n,xscale=:log10,yscale=:log10,ylabel="Varianza de la varianza (escala log)", 
    xlabel="log(n)",label=false,color="#13165B", 
        marker=:star,title="Varianza de la varianza en funciÃ³n del nÃºmero de pasos n.")
```
