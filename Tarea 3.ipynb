```julia
#Creamos una función que genere una realización de caminante de n elementos 
function realizacion(n)
    #La lista donde se anexarón los pasos 
   lista=[]
    #Un ciclo que cree n pasos de la realización
    for i in 1:n
        push!(lista,rand((-1,1)))
    end
    #Se regresa la lista con los pasos
    return lista
end
```

Construyendo las L listas para n=10,100,1000,10000

```julia
#Las variables: los elementos que puede tomar n
n=[10,100,1000,10000]
#El número de listas para cada n
L=10000
#Creamos una lista donde anexaremos las realizaciones
Listas_4=[]
#Para saber a que elemento de n corresponde la lista_4-i creamos un arreglo de ns
ns=[]
#Para cada n dada creamos 10000 listas:
for i in n
    for j in 1:L
        #anexamos esas listas a Listas_4 y anexamos también la n correspondiente a cada n 
        push!(Listas_4,realizacion(i))
        push!(ns,i)
    end
end
```

Creamos una función que permita obtener el promedio de una lista:

```julia
function promedio(lista)
    #variable de suma
    suma=0
    #Ciclo que recorra todos los elementos de la lista
    for i in 1:length(lista)
        #Suma el i-esimo elemento de la lista a la variable suma
        suma+=lista[i]
    end
    #Dividimos la suma entre el número de elementos para obtener el promedio y regresa este valor
    return suma/length(lista)
end
```

```julia
#Obteniendo el promedio de cada lista
promedios_4=[]
for i in Listas_4
    push!(promedios_4,promedio(i))
end
```

```julia
#Agrupamos en listas las listas de realizaciones conforme a su debida n
listas_10=[]
listas_100=[]
listas_1000=[]
listas_10000=[]
for i in 1:length(Listas_4)
    if ns[i]==10
        push!(listas_10,Listas_4[i])
    end
    if ns[i]==100
        push!(listas_100,Listas_4[i])
    end
    if ns[i]==1000
        push!(listas_1000,Listas_4[i])
    end
    if ns[i]==10000
        push!(listas_10000,Listas_4[i])
    end
end
```

a. Obteniendo el promedio de los promedios, para ello creamos una función que obtenga el promedio de los promedios:

```julia
function promedio_promedios(lista)
    #Creamos una lista donde almacenaremos los promedios del promedio
    promedios=[]
    for i in 1:length(lista)
        #obtenemos el promedio de cada lista y lo almacenamos en la lista promedios
        push!(promedios,mean(lista[i]))
    end
    #Obtenemos el promedio de los promedios y lo regresamos con la función
return mean(promedios)
end
```

Presentamos los promedios en una tabla para mejor visualizacion

```julia
#Obteniendo los promedios de los promedios de cada lista que contiene las L realizaciones de n=10,100,1000,10000 pasos
promedio_promedio_10=promedio_promedios(listas_10)
promedio_promedio_100=promedio_promedios(listas_100)
promedio_promedio_1000=promedio_promedios(listas_1000)
promedio_promedio_10000=promedio_promedios(listas_10000)
DataFrame(n10=promedio_promedio_10,n100=promedio_promedio_100,
                m1000=promedio_promedio_1000,n10000=promedio_promedio_10000)
```

b. Calculando las varianzas de la varianza usando el mismo procedimiento que para los promedios de los promedios

```julia
#Creamos la función que nos cree la varianza de la varianza de una lista
function varianza_promedio(lista)
    #Creamos una variable que almacenará la vrianza de la iesima realización de la lista
    promedio=[]
    for i in 1:length(lista)
        push!(promedio,mean(lista[i]))
    end
    #Regresamos la varianza de las varianza
return var(promedio)
end
```

Calculando la varianza de las varianzas

```julia
#Obteniendo la varianza de las varianzas
varianza_varianza_10=varianza_promedio(listas_10)
varianza_varianza_100=varianza_promedio(listas_100)
varianza_varianza_1000=varianza_promedio(listas_1000)
varianza_varianza_10000=varianza_promedio(listas_10000)
DataFrame(n10=varianza_varianza_10,n100=varianza_varianza_100,
                m1000=varianza_varianza_1000,n10000=varianza_varianza_10000)
```

Graficando la varianza de las varianzas en escala logaritmica en ambos ejes.

```julia
varianza_varianza_n=[varianza_varianza_10,varianza_varianza_100,varianza_varianza_1000,varianza_varianza_10000]
plot(n,varianza_varianza_n,xscale=:log10,yscale=:log10,ylabel="Varianza de la varianza (escala log)", 
    xlabel="log(n)",label=false,color="#13165B", 
        marker=:star,title="Varianza del promedio en función del número de pasos n.")
```
